```python
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
url = 'https://www.furusato-tax.jp/rank?header'
res = requests.get(url)
res.status_code

if res.status_code == 200 :
    print(res)
    print('response success!!')
```

```python
html_doc = res.text
print(html_doc[0:30000])
```

```python
# html.parserでHTMLの構造を解析
soup = BeautifulSoup(html_doc, 'html.parser')
# htmlをインデント等のフォーマットされた状態で出力 
print(soup.prettify())
```

```python
# liタグの取得
soup.find_all('li')
# soup.find_all('li')[0]でリスト0番目のタグを取得
# soup.li でも取得可能
# soup.li.textでタグ内のテキストを取得
```

```python
# 範囲を選択して取得
# rﾝキングを取得
rankings = soup.find_all('li', attrs={'class':'grid__block sld__item'})[0]
rankings
```

```python
# ランキングの中から品目を取得
rank_titles = rankings.find('h3', attrs={'class':'card-product__title'})
rank_titles.text
```

```python
# 価格取得
prices = rankings.find_all('p', attrs={'class':'card-product__price'})[0]
prices.text.split('x')
prices.text.replace('x','')
# int(~~)で整数型に変換
```

```python
# 市町村取得
cities = rankings.find_all('a', attrs={'class':'card-product__city'})
cities
```

```python
city = cities[0].span
city
```

```python
data = []

rankings = soup.find_all('li', attrs={'class':'grid__block sld__item'})
for ranking in rankings:
    rank_titles = ranking.find('h3', attrs={'class':'card-product__title'})
    prices = ranking.find('p', attrs={'class':'card-product__price'})
    cities = ranking.find('a', attrs={'class':'card-product__city'})

    print(rank_titles.text)
    print(prices.text)
    print(city.text)
    print('---------------------------')
    data.append(rank_titles.text)
    data.append(prices.text)
    data.append(city.text)
```
