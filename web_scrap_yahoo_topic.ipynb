```python
# Webデータの取得
# yahooトップニュースの項目取得
```

```python
import requests
from time import sleep
```

```python
url = 'https://yahoo.co.jp'
res = requests.get(url)
res.status_code
print(res)
```

```python
html_doc = res.text
print(html_doc[7:200])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser') #html.parser:HTMLの構造を解析するもの
```

```python
soup
```

```python
print(soup.prettify())
```

```python
p_0 = soup.find_all('p')[0] #1番目のpタグを取得
print(p_0)
# リストのような形式になっているので、pタグすべて取得し、リスト番号を選択することが可能
```

```python
t_all = soup.find_all('title') #find_allでタグの物を全て取得
print(t_all)
```

```python
# タグ名指定でも取得可能
soup.title
# タグ名＋textで中身だけ抽出
soup.title.text
```

```python
# 情報を取得する範囲を絞る。タグを指定し、attrs引数で絞る(classが_3cl937~のものを取得してくる)
soup.find_all('h1', attrs={'class' : '_3cl937Zpn1ce8mDKd5kp7u'})
```

```python
# リスト形式の中身をテキストで出力
top_topic = soup.find_all('h1', attrs={'class' : '_3cl937Zpn1ce8mDKd5kp7u'})[0]
top_topic.text
```

```python
top_topic = soup.find('h1', class_= '_3cl937Zpn1ce8mDKd5kp7u')
top_topic.get_text()
```

```python
# 情報を取得する範囲を絞る。
top_topic_all = soup.find_all('h1', class_= '_3cl937Zpn1ce8mDKd5kp7u')  
top_topic_all[1]
```

```python
for topic in top_topic_all:
    print(topic)
    sleep(2) # 間隔をあけて表示させる
```
