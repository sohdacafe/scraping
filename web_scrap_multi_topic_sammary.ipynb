```python
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
url = 'https://industrial.panasonic.com/jp/products/pt/lithium-ion/models'
res = requests.get(url)
print(res)
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# １つの機種の情報取得
```

```python
odd_category_items = soup.find_all('tr', attrs={'class': 'odd'})
odd_category_item = odd_category_items[0]
odd_category_item
```

```python
# 更に<td class="model nowrap"><a href="/jp/products/pt/lithium-ion/models/NCA103450">NCA103450</a><br/></td> からモデル名を探す
```

```python
model_name = odd_category_item.find('td', attrs={'class' : 'model nowrap'})
model_name
```

```python
model_name.text
```

```python
# 他の項目探索
```
